# Processing County by Population

This notebook is for processing the counties included in the OSPI dataset by population in order to create useful feature(s).

```python
import pandas as pd
import numpy as np
```

```python
original_ospi = pd.read_csv('2022_school_pt5.csv',  index_col=0)
```

```python
census_df = pd.read_csv('co-est2023-alldata.csv',  index_col=0, encoding = 'latin')
```

```python
census_df.head()
```

```python
wa_counties = census_df.loc[census_df['STNAME'] == 'Washington'].copy(deep=True)
```

```python
wa_counties.head()
```

```python
# wa_pop = wa_counties[['POPESTIMATE2023','CTYNAME']].head(50).copy(deep=True)
wa_pop = wa_counties.loc[:, ('POPESTIMATE2023', 'CTYNAME')]
```

```python
wa_pop = wa_pop[((wa_pop.CTYNAME != 'Washington'))].copy(deep=True)
```

```python
string = "Whatcom County"
string[:-7]
```

```python
def remove_county(row):
    return row['CTYNAME'][:-7]

wa_pop['County'] = wa_pop.apply(remove_county, axis=1)
```

```python
wa_pop = wa_pop.loc[:, ('POPESTIMATE2023', 'County')]
```

```python
wa_pop.head()
```

```python
ospi_with_pop = pd.merge(wa_pop, original_ospi, how='outer', on='County')
```
